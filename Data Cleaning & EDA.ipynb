# Cleaning Data

```python
# import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read files

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
df
```

```python
# check null values
df.isnull().sum()
```

```python
# check nan values

df.isna().sum()
```

```python
# check duplicate values

df2 = df.copy()
df2['duplicated'] = df2.duplicated()
df2
```

```python
# print index contains duplicate values

print(df2[df2['duplicated']]==True)
```

```python
df.describe()
```

```python
df.info()
```

```python
# check categorical values

df['Age'].unique()
```

```python
# function for age categories

def ages(angka):
    if 26 <= angka <= 35:
        return "26-35"
    elif 36 <= angka <= 45:
        return "36-45"
    elif 46 <= angka <= 56:
        return "46-56"
    else:
        return "56-59"
```

```python
# applying function ages

df['Age_c'] = df['Age'].apply(ages)
df
```

```python
# check categorical values

df['Gender'].unique()
```

```python
# check categorical values

df['Occupation'].unique()
```

```python
# aligning same categorical values

df['Occupation'] = df['Occupation'].replace(['Software Engineer','Sales Representative'], ['Engineer','Salesperson'])
df['Occupation'].unique()
```

```python
# check categorical values

df['BMI Category'].unique()
```

```python
# aligning same categorical values

df['BMI Category'] = df['BMI Category'].replace(['Normal Weight'], 'Normal')
df['BMI Category'].unique()
```

```python
# check categorical values

df['Blood Pressure'].unique()
```

```python
# check categorical values

df['Sleep Disorder'].unique()
```

```python
# check columns
df.columns
```

```python
# drop unused columns

df.drop(columns=['Person ID'])
```

```python
# Feature engineering for stress level

df['stress'] = df['Stress Level'].apply(lambda x: 1 if x >= 5 else 0)
df
```

# Exploratory Data Analysis

```python
# checking distribution of each numerical columns

df.hist(bins=60, figsize=(20,10))
```

```python
# Checking correlation between numerical columns using pearson correlation

corr = df.corr().round(2)
plt.figure(figsize = (20,10))
sns.heatmap(corr, annot = True)
```

### PLOTTING

```python
# checking the frequency of each occupation

df['Occupation'].value_counts().sort_index().plot(kind='bar')
```

```python
# checking the distribution of each occupation

df['Age_c'].value_counts().sort_index().plot(kind='bar')
```

```python
num_col=['Sleep Duration',
       'Quality of Sleep', 'Physical Activity Level', 'Stress Level',
       'Heart Rate', 'Daily Steps']
```

```python
# check data distribution with age category

fig = plt.figure(figsize=(15,8))
for i in range(len(num_col)):
    plt.subplot(3,3,i+1)
    plt.title(num_col[i])
    sns.boxplot(data=df,y=df['Age_c'],x=df[num_col[i]])
plt.tight_layout()
plt.show()
```

```python
# check data distribution with Occupation category

fig = plt.figure(figsize=(15,8))
for i in range(len(num_col)):
    plt.subplot(3,3,i+1)
    plt.title(num_col[i])
    sns.boxplot(data=df,y=df['Occupation'],x=df[num_col[i]])
plt.tight_layout()
plt.show()
```

## Since the distribution data fits with the age category, the next analysis will be compare with age variables
